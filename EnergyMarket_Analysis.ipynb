## Data Cleaning

#### Changing all the date/time column from object d-type to date-time d-type

```python
# For dd/mm/yyyy format
historical_data['Date/Time'] = pd.to_datetime(
    historical_data['Date/Time'],
    format="%d/%m/%Y",
    errors='coerce'
)
infrastructure_data['Date/Time'] = pd.to_datetime(
    infrastructure_data['Date/Time'],
    format="%d/%m/%Y",
    errors='coerce'
)

# For yyyy-mm-dd format
market_data['Date/Time'] = pd.to_datetime(
    market_data['Date/Time'],
    format="%Y-%m-%d",
    errors='coerce'
)
regulatory_data['Date/Time'] = pd.to_datetime(
    regulatory_data['Date/Time'],
    format="%Y-%m-%d",
    errors='coerce'
)
```

#### Aggregating and resampling historical dataset

```python
historical_data.head()
```

```python
hist_daily_total = (historical_data.groupby(['Date/Time', 'Location/Region', 'Energy Source'], as_index = False).agg({'Energy Production (kWh)': 'sum', 'Energy Consumption (kWh)': 'sum', 'Energy Demand': 'sum', 'Energy Price': 'mean', 'Operational Costs': 'sum' }))
```

#### Merging all the dataset into one

```python
merged = hist_daily_total.merge(
    market_data,
    on = 'Date/Time',
    how = 'left',
    suffixes = ('','_market'))
```

```python
merged = merged.merge(
    infrastructure_data,
    on = 'Date/Time',
    how = 'left',
    suffixes = ('','_infra'))
```

```python
merged = merged.merge(
    regulatory_data,
    on = 'Date/Time',
    how = 'left',
    suffixes = ('','_reg'))
```

```python
merged.head()
```

#### There are 13 missing fields in 'Technology_limitations' field so filling them with 'Unknown'

```python
merged['Technology Limitations'] = merged['Technology Limitations'].fillna('Unknown')
```

#### Checking missing values

## KPI Calculations

### 1. Production Efficiency
merged['Production Efficiency'] = (
    merged['Energy Production (kWh)'] / merged['Energy Demand (kWh)']
)

### 2. Cost per kWh
merged['Cost per kWh'] = (
    merged['Operational Costs (USD)'] / merged['Energy Production (kWh)']
)

### 3. Price Competitiveness
merged['Price Competitiveness'] = (
    merged['Energy Price (USD/kWh)'] - merged['Market Price (USD/kWh)']
)

### 4. Demand Gap
merged['Demand Gap'] = (
    merged['Market Demand (kWh)'] - merged['Energy Demand (kWh)']
)

### 5. Tech Limitation Score (map categorical to numeric risk score)
risk_map = {
    'None': 0,
    'Low': 1,
    'Moderate': 2,
    'High': 3,
    'Unknown': None
}
merged['Tech Limitation Score'] = merged['Technology Limitations'].map(risk_map)

### 6. Compliance Cost Ratio
merged['Compliance Cost Ratio'] = (
    merged['Compliance Costs (USD)'] / merged['Operational Costs (USD)']
)

### 7. Regulation Impact Days (boolean flag: 1 if any change reported, else 0)
merged['Regulation Impact'] = merged['Regulatory Changes'].notna().astype(int)



```python
merged['Production Efficiency'] = merged['Energy Production (kWh)'] / merged['Energy Demand'].round(2)

merged['Cost per kWh'] = ( merged['Operational Costs'] / merged['Energy Production (kWh)'].round(2) )

merged['Price Competitiveness'] = ( merged['Energy Price'] - merged['Market Price'].round(2) )

merged['Demand Gap'] = ( merged['Energy Demand'] - merged['Energy Production (kWh)'])

risk_map = { 'None': 0, 'Low': 1, 'Moderate': 2, 'High': 3, 'Unknown': None } 
merged['Tech Limitation Score'] = merged['Technology Limitations'].map(risk_map)

merged['Compliance Cost Ratio'] = merged['Compliance Costs'] / merged['Operational Costs'].round(2) 

merged['Regulation Impact'] = (merged['Regulatory Changes'] != 'No change').astype(int)

# 8. Renewable Energy Share (%)
# First, calculate total energy production from all sources.
total_production = merged['Energy Production (kWh)'].sum()

# Next, calculate the total energy production from renewable sources.
renewable_production = merged[merged['Energy Source'] == 'Renewables']['Energy Production (kWh)'].sum()

# Calculate the share as a percentage.
merged[renewable_energy_share] = (renewable_production / total_production) * 100

# Check the first few rows to verify the new KPI columns
print("First 5 rows of the merged dataframe with new KPIs:")
print(merged.head())
print("\n--- Summary KPIs ---")
print(f"Renewable Energy Share: {renewable_energy_share:.2f}%")
```

## Ploting and Visualisation

```python
# 1. Plot for Demand Fluctuations (Business Problem A)
# Line chart to show how the Demand Gap changes over time
plt.figure(figsize=(12, 6))
plt.plot(merged['Date/Time'], merged['Demand Gap'], label='Demand Gap (kWh)', color='royalblue')
plt.axhline(0, color='red', linestyle='--', label='Zero Demand Gap')
plt.title('Demand Gap Over Time')
plt.xlabel('Date')
plt.ylabel('Demand Gap (kWh)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('demand_gap_plot.png')
print("Generated 'demand_gap_plot.png'")
plt.close()
```

```python
# 3. Plot for Regulatory Changes (Business Problem C)
# Line plot of Compliance Costs with markers for regulatory changes
plt.figure(figsize=(12, 6))
plt.plot(merged['Date/Time'], merged['Compliance Costs'], label='Compliance Costs', color='darkorange')
# Plot markers only on days with a regulation impact
regulation_days = merged[merged['Regulation Impact'] == 1]
plt.scatter(regulation_days['Date/Time'], regulation_days['Compliance Costs'], color='red', zorder=5, label='Regulation Change')
plt.title('Compliance Costs with Regulation Impact')
plt.xlabel('Date')
plt.ylabel('Compliance Costs (USD)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('compliance_cost_plot.png')
print("Generated 'compliance_cost_plot.png'")
plt.close()
```

```python
# 6. New Plot: Regulation Impact vs. Production Efficiency
# Prepare data for the box plot
data_to_plot_efficiency = [
    merged[merged['Regulation Impact'] == 0]['Production Efficiency'], # Days without regulation impact
    merged[merged['Regulation Impact'] == 1]['Production Efficiency']  # Days with regulation impact
]
labels_efficiency = ['No Regulation Impact', 'Regulation Impact']

plt.figure(figsize=(8, 6))
plt.boxplot(data_to_plot_efficiency, patch_artist=True, labels=labels_efficiency,
            boxprops=dict(facecolor='lightgreen', color='green'),
            medianprops=dict(color='red'))
plt.title('Production Efficiency Distribution with Regulation Impact')
plt.xlabel('Regulatory Status')
plt.ylabel('Production Efficiency')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('efficiency_vs_regulation.png')
print("Generated 'efficiency_vs_regulation.png'")
plt.close()
```
